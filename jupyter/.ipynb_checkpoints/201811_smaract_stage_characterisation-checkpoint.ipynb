```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import sys,os
import glob
sys.path.append('/data/id13/inhouse2/AJ/skript/')
import simplecalc.fitting2d as fit2d
import fileIO.datafiles.open_data as odata
```

testing the circle fitting used later to remove COR error from data

```python
phi = np.linspace(0.,360.,360)
x = np.zeros_like(phi)
y = np.zeros_like(phi)
x[0:90]=y[0:180]=1
x[90:180]=y[180:270] = np.linspace(1.,-1.,90)
x[180:270]=y[270:360] = -1
x[270:360]=y[0:90] = np.linspace(-1.,1.,90)


res = fit2d.do_fit_circle_in_2d_data(phi,x=x,y=y,verbose=True)
print('COR [x,y]: [{},{}]'.format(res[0][0],res[0][1]))
```

reading the data into a dict


```python
small_data={}
med_data = {}
mean_data = {}
path ='/mntdirect/_data_id13_inhouse2/AJ/data/20181107_smaract_rotation_stages/stages_characterisation/*'

small_data_calibration = 495./5000  # calibration was 495 pxl/mm should have been 5000
med_data_calibration   =   2./109.*38. # calibration was 495 pxl/mm should have been 495/2, 109/38 x longer lever

dat_fname_list_med = [x for x in glob.glob(path) if x.find('long_')>-1]
dat_fname_list_small = [x for x in glob.glob(path) if x.find('series_')>-1]
for i,fname in enumerate(dat_fname_list_small):        
    data, header = odata.open_data(fname)
    data=data.T
    small_data[i]=dict(zip(header,[x for x in data]))
    small_data[i]['x'] = x = small_data[i]['x'] *small_data_calibration
    small_data[i]['y'] = y = small_data[i]['y'] *small_data_calibration
    if 'small' not in mean_data.keys():
        mean_data['small'] = {}
        mean_data['small']['phi'] = np.zeros_like(small_data[i]['phi'])
        mean_data['small']['x'] = np.zeros_like(small_data[i]['x'])
        mean_data['small']['y'] = np.zeros_like(small_data[i]['y'])
    mean_data['small']['phi'] += small_data[i]['phi']
    mean_data['small']['x'] += x
    mean_data['small']['y'] += y

for value in mean_data['small'].values()[:]:
    value *= 1./(i+1)
    
for i,fname in enumerate(dat_fname_list_med):
    data, header = odata.open_data(fname)
    data=data.T
    med_data[i]=dict(zip(header,[x for x in data]))

    med_data[i]['x'] = x = med_data[i]['x'] *med_data_calibration
    med_data[i]['y'] = y = med_data[i]['y'] *med_data_calibration
    if 'med' not in mean_data.keys():
        mean_data['med'] = {}
        mean_data['med']['phi'] = np.zeros_like(med_data[i]['phi'])
        mean_data['med']['x'] = np.zeros_like(med_data[i]['x'])
        mean_data['med']['y'] = np.zeros_like(med_data[i]['y'])
    
    mean_data['med']['phi'] += med_data[i]['phi']
    mean_data['med']['x'] += x
    mean_data['med']['y'] += y 
for value in mean_data['med'].values()[:]:
    value *= 1./(1+i)      

data_dict = {}
data_dict.update({'small':small_data})
data_dict.update({'med':med_data})
[plt.plot(small_data[i]['phi'],small_data[i]['y']) for i in range(10)] 
[plt.plot(small_data[i]['phi'],small_data[i]['x']) for i in range(10)]
[plt.plot(med_data[i]['phi'],med_data[i]['y']) for i in range(10)] 
[plt.plot(med_data[i]['phi'],med_data[i]['x']) for i in range(10)]
```

```python
fit_data={}
for key, data in mean_data.items():
    circle_res = fit2d.do_fit_circle_in_2d_data(data['phi'],data['x'],data['y'],verbose=False)
    fit_data[key]={}
    fit_data[key]['phi']=data['phi']
    x,y = fit2d.circle_func(circle_res[0],data['phi'])
    fit_data[key]['x'] = x
    fit_data[key]['y'] = y
    plt.plot(data['x'],data['y'],'bo')
    plt.plot(fit_data[key]['x'],fit_data[key]['y'],'r--')
```

Subtract the fit to the mean data from each individual dataset (residual)

```python
fig,ax = plt.subplots()
ax.set_xlabel('x posistion in [mm]')
ax.set_ylabel('y posistion in [mm]')
analysed_dict = data_dict['analysed'] = {}
for key in ['small','med']:
    data = data_dict[key]
    corr_dict = analysed_dict[key] = {}

    for measurement_no, measurement in data.items():
        
        corr_dict[measurement_no] = {}
        corr_dict[measurement_no]['phi'] = measurement['phi']
        x = corr_dict[measurement_no]['x'] = measurement['x'] - fit_data[key]['x']
        y = corr_dict[measurement_no]['y'] = measurement['y'] - fit_data[key]['y']
        ax.plot(x,y)
    
```

```python
data_dict['circle_fit'] = {}
for key in ['small','med']:
    data = analysed_dict[key]
    
    cf = data_dict['circle_fit'][key] = {}
    
    for measurement_no , measurement in data.items():
        circle_res = fit2d.do_fit_circle_in_2d_data(measurement['phi'],measurement['x'],measurement['y'],verbose=False)
        cf[measurement_no] = {}
        cf[measurement_no]['fit_result'] = circle_res[0]
        x,y = fit2d.circle_func(circle_res[0],measurement['phi'])
        cf[measurement_no]['residual_x'] = data_dict[key][measurement_no]['x'] -  mean_data[key]['x'] - x
        cf[measurement_no]['residual_y'] = data_dict[key][measurement_no]['y'] -  mean_data[key]['y'] - y
        
    
```

```python
fig,ax = plt.subplots()
[ax.plot(data_dict['circle_fit']['small'][x]['fit_result'][0],data_dict['circle_fit']['small'][x]['fit_result'][1],'rx') for x in range(10)]
[ax.plot(data_dict['circle_fit']['med'][x]['fit_result'][0],data_dict['circle_fit']['med'][x]['fit_result'][1],'bo') for x in range(10)]
ax.set_xlabel('optimal COR x position [mm]')
ax.set_ylabel('optimal COR y position [mm]')
ax.set_xlim(-0.002,0.002)
ax.set_ylim(-0.002,0.002)
```

```python
fig,ax = plt.subplots()
[ax.plot(data_dict['small'][x]['phi'],data_dict['circle_fit']['small'][x]['residual_x'],'rx') for x in range(10)]
[ax.plot(data_dict['small'][x]['phi'],data_dict['circle_fit']['small'][x]['residual_y'],'bo') for x in range(10)]
ax.set_ylabel('residual [mm]')
ax.set_xlabel('phi')
```
