```python
import sys,os
import numpy as np
sys.path.append('/data/id13/inhouse2/AJ/skript')
import fileIO.images.image_tools as it
import cv2
import matplotlib.pyplot as plt
```

```python
# flat frame

FPS = 60
FREQ = 14
VID_TIME = 240

FRAME_WIDTH = 48
FRAME_HEIGHT = 27
FRAME_SHAPE = (FRAME_WIDTH, FRAME_HEIGHT)

NO_FRAMES = VID_TIME * FPS
SPATH = '/data/id13/inhouse2/AJ/data/ma3576/test/'
SAVE_FNAME = SPATH + 'fps_{:03d}_freq_sweep.avi'.format(FPS,FREQ)
if not os.path.exists(SPATH):
    os.mkdir(SPATH)
w,h,d = np.meshgrid(np.linspace(0,1,FRAME_WIDTH),np.linspace(0,1,FRAME_HEIGHT),np.arange(3))
writer = cv2.VideoWriter(SAVE_FNAME, cv2.VideoWriter_fourcc(*"MJPG"), FPS, FRAME_SHAPE)
frame = np.ones((FRAME_HEIGHT,FRAME_WIDTH,3),dtype=np.uint8)
for i in range(NO_FRAMES):
    if i<0.5*NO_FRAMES:
        sweep = 0.1 + 1.8*(float(i)/ NO_FRAMES)
    else:
        sweep = 0.1 + 1.8*(float(NO_FRAMES - i)/ NO_FRAMES)
    
    phase = 0.5*(np.sin(float(i)*FREQ*sweep/FPS*1*np.pi)+1)
    phase2 = 0.5*(np.sin(float(0.123*i)*FREQ/FPS*1*np.pi)+1)
    weight = np.int(phase*255)
    # side_func = 0.5*(np.sin((w+phase2)*np.pi*2)+1)
    weights = np.asarray(weight*frame,np.uint8)
    # weights = np.asarray(weight*side_func,np.uint8)
    print('recording frame {} of {}, weight {:2.2f}'.format(i+1, NO_FRAMES, sweep))
    writer.write(weights)
writer.release()
```

```python
# gauss

FPS = 60
FREQ = 12
VID_TIME = 240

FRAME_WIDTH = 48
FRAME_HEIGHT = 27
FRAME_SHAPE = (FRAME_WIDTH, FRAME_HEIGHT)

NO_FRAMES = VID_TIME * FPS
SPATH = '/data/id13/inhouse2/AJ/data/ma3576/test/'
SAVE_FNAME = SPATH + 'fps_{:03d}_freq_sweep_gauss4.avi'.format(FPS,FREQ)

X,Y=np.meshgrid(np.linspace(-1,1,FRAME_WIDTH),np.linspace(-1,1,FRAME_HEIGHT));
mu,sigma=0,0.5; #suppose that mux=muy=mu=0 and sigmax=sigmay=sigma
G=np.exp(-((X-mu)**2+(Y-mu)**2)/2.0/sigma**2)

if not os.path.exists(SPATH):
    os.mkdir(SPATH)
w,h,d = np.meshgrid(np.linspace(0,1,FRAME_WIDTH),np.linspace(0,1,FRAME_HEIGHT),np.arange(3))
writer = cv2.VideoWriter(SAVE_FNAME, cv2.VideoWriter_fourcc(*"MJPG"), FPS, FRAME_SHAPE)

white = np.ones((FRAME_HEIGHT,FRAME_WIDTH,3))
gauss = np.dstack([G,G,G])
for i in range(NO_FRAMES):
    if i < (0.5*NO_FRAMES):
        sweep = 0.1 + 1.8*(float(i)/ NO_FRAMES)
    else:
        sweep = 0.1 + 1.8*(float(NO_FRAMES - i)/NO_FRAMES)
    
    phase = 0.5*(np.sin(float(i)*FREQ*sweep/FPS*1*np.pi)+1)
    phase2 = (np.sin(float(i)*0.97*sweep*FREQ/FPS*1*np.pi))
    
    phase_2d = (white*phase + gauss*phase2)
    weights = 255.*np.sin(phase_2d)
                        
    frame = np.asarray(weights,dtype=np.uint8)
    print('recording frame {} of {}, weight {:2.2f}'.format(i+1, NO_FRAMES, weights.max()))
    writer.write(frame)
writer.release()
```
